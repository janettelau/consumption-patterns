# Import the libraries

# Define paths and read files

```python
# Store filepath in a variable
file_one = Path("Resources/Cost_Affordability_healthy_Diet_Data.csv")
file_two = Path("Resources/Obesity_unhealthy_diet.csv")
file_3 = Path("Resources/Raw_WIDER_data.csv")
```

```python
# Read our data file with the Pandas library
# Not every CSV requires an encoding, but be aware this can come up
coahd_df = pd.read_csv(file_one, encoding="ISO-8859-1")
obesity_df = pd.read_csv(file_two, encoding="ISO-8859-1")
income_df = pd.read_csv(file_3, encoding="ISO-8859-1")
```

# Cleaning Cost and Affordability of a Healthy Diet (CoAHD) data

```python
# Show the rows and colomns of each dataframes
coahd_df # Cost and Affordability of a Healthy Diet (CoAHD)
coahd_df.head()
```

```python
coahd_df.describe()
```

```python
# Get a list of columns
cols = coahd_df.columns
cols
```

```python
#Create a new df with desired columns
new_coahd_df =coahd_df[['Area Code', 'Area', 'Item Code', 'Item',
       'Element Code', 'Element', 'Unit', 'Y2017',  'Y2018', 
       'Y2019',  'Y2020',  'Y2021',  'Y2022',
       ]]
new_coahd_df.head(12)
```

```python
# Verifying for columns with NaN
columns_with_nan = new_coahd_df.columns[new_coahd_df.isna().any()].tolist()
columns_with_nan
```

```python
new_coahd_df.dtypes
```

```python
new_coahd_df.loc[:, "Y2017"] = new_coahd_df.loc[:, "Y2017"].replace('<0.1', '0.09')
```

```python
new_coahd_df.loc[ : ,['Y2017', 'Y2018', 'Y2019', 'Y2020', 'Y2021', 'Y2022']] = new_coahd_df.loc[ :,['Y2017', 'Y2018', 'Y2019', 'Y2020', 'Y2021', 'Y2022']].replace('<0.1',  "0.09")
```

```python
coahd_df_new = new_coahd_df.astype({'Y2017' : float, 
                                    'Y2018' : float, 
                                    'Y2019' : float, 
                                    'Y2020': float, 
                                    'Y2021' : float, 
                                    'Y2022' : float
                                    })
```

```python
coahd_df_new.dtypes
```

```python
# Replacing the NaN value with 0 or '' depending on type of columns
coahd_wo_NaN_df = coahd_df_new.fillna( {'Y2017' :0 , 
                                    'Y2018' : 0, 
                                    'Y2019' : 0, 
                                    'Y2020': 0, 
                                    'Y2021' : 0, 
                                    'Y2022' : 0})
coahd_wo_NaN_df.head(30)
```

```python
# Verifying for columns with NaN
columns_with_nan = coahd_wo_NaN_df.columns[coahd_wo_NaN_df.isna().any()].tolist()
columns_with_nan
```

```python
coahd_cleaned_df = coahd_wo_NaN_df
coahd_cleaned_df
```
