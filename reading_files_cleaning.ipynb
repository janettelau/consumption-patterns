# Import the libraries

```python
# Dependencies
import pandas as pd
from pathlib import Path
```

# Define paths and read files

```python
# Store filepath in a variable
file_one = Path("Resources/Cost_Affordability_healthy_Diet_Data.csv")
file_two = Path("Resources/Obesity_unhealthy_diet.csv")
file_3 = Path("Resources/Raw_WIDER_data.csv")
```

```python
# Read our data file with the Pandas library
# Not every CSV requires an encoding, but be aware this can come up
coahd_df = pd.read_csv(file_one, encoding="ISO-8859-1")
obesity_df = pd.read_csv(file_two, encoding="ISO-8859-1")
income_df = pd.read_csv(file_3, encoding="ISO-8859-1")
```

# Cleaning Global income statistics data

```python
# Show the rows and colomns of each dataframe
income_df.head() # Global income statistics
```

```python
income_df.describe()
```

```python
cols = income_df.columns
cols
```

```python
# Keep only desired columns ( from 36 to 20)
global_income_df = income_df[[ 'country', 'year', 
       'rural_pop_pct', 'gini_reported', 'q1', 'q2', 'q3', 'q4', 'q5',   
        'region_un',  
       'oecd', 'incomegroup',  
         'gdp_ppp_pc_usd2011',
       'population', 'rural_pop', 'urban_pop', 
       'staple_demand']]
```

```python
global_income_df.head()
```

```python
# Verifying for columns with NaN
columns_with_nan = global_income_df.columns[global_income_df.isna().any()].tolist()
columns_with_nan
```

```python
# Replacing the NaN value with 0 or '' depending on type of columns
global_inc_wo_nan_df = global_income_df.fillna( {"rural_pop_pct" : 0,
                                                'gini_reported': 0,
                                                'q1' : 0,
                                                'q2' : 0,
                                                'q3' : 0,
                                                'q4' : 0,
                                                'q5' : 0,
                                                'pop_gdp' : 0,
                                                'population' : 0,
                                                'staple_demand' : 0   })
global_inc_wo_nan_df
```

```python
global_inc_wo_nan_df.count()
```

```python
# Describe the dataframe for # Global income statistics

global_inc_wo_nan_df.describe()
```

```python
# Renaming some columns
global_income_cleaned_df= global_inc_wo_nan_df.rename(columns = {'gdp_ppp_pc_usd2011' : "pop_gdp",
                                                      "oecd": "oecd_member",
                                                       "areacovr": "area_covered",
                                                       "population" : "population count"})
global_income_cleaned_df
```

```python
# Get the data types for each column
global_income_cleaned_df.dtypes
```

```python
# Get the unique income groups
global_income_cleaned_df['incomegroup'].unique()
```

```python
# Get the unique countries
global_income_cleaned_df['country'].unique()
```

```python
# Get the number of unique countries
global_income_cleaned_df['country'].nunique()
```

```python
# Get the unique years
global_income_cleaned_df['year'].unique()
```

```python
# Keep only years after 2010
global_income_cleaned_new_df = global_income_cleaned_df.loc[global_income_cleaned_df['year'] >= 2010, : ]
global_income_cleaned_new_df
```

```python
# Keep only year 2017
global_income_cleaned_new_2017_df = global_income_cleaned_df.loc[global_income_cleaned_df['year'] == 2017, : ]
global_income_cleaned_new_2017_df
```

```python
# Get list of remaining unique countries
global_income_cleaned_new_2017_df['country'].unique()
```

```python
# Get the number of remaining Unique countries
global_income_cleaned_new_2017_df['country'].nunique()
```

```python
cols_list = global_income_cleaned_new_2017_df.columns
cols_list
```

```python
global_inc_object_cols_df=global_income_cleaned_new_2017_df [['country','region_un', 'oecd_member', 'incomegroup']]
global_inc_object_cols_df
```

```python
# Group by 'Country' and calculate the average for all numeric columns
country_aggregated = global_income_cleaned_new_2017_df.groupby('country').agg('mean', numeric_only=True)
country_aggregated
```

```python
# Include all object columns in the dataframe
global_inc_cleaned_final_df = pd.merge(country_aggregated, global_inc_object_cols_df, on="country", how= 'inner' )
global_inc_cleaned_final_df

```

```python
# Drop the duplicate rows
unique_df = global_inc_cleaned_final_df.drop_duplicates(subset='country')
unique_df
```

```python
# Change the data type for year to int
unique_df.loc[ : ,'year'] = unique_df.loc[ : , 'year' ].astype(int)
unique_df
```

```python
# Change the data type for year to str
unique_df.loc[ : ,'year'] = unique_df.loc[ : , 'year' ].astype('str')
unique_df
```

```python
# Renaming the final global income data frame
global_income_final_df = unique_df
global_income_final_df
```

```python
country_list = global_income_final_df['country'].unique()
country_list
```

```python
global_income_final_df['country'].nunique()
```
