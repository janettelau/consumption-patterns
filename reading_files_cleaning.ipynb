```python
# Create unique item  list 
import re
# List of keywords that indicate categories
category_keywords = [
    'Developed',
    'Developing',
    'Income',
    'Countries',
    'economies',
    'States',
    'Food Deficit',
    'Least Developed',
    'Land Locked',
    'Africa',
    'ern America',
    'World'
    'Asia',
    'Europe'
]
# Function to determine if an entry is likely a category
def is_category(area):
    return any(keyword in area for keyword in category_keywords)
# Function to determine if an entry is a country
def is_country(area):
    # Check for patterns that are typical of country names
    return bool(re.match(r'^[A-Z][a-zA-Z\s()\'-]+$', area) and not is_category(area))

# Create new columns for Country and Category
coahd_cleaned_df_new = coahd_cleaned_df
coahd_cleaned_df_new['Country'] = coahd_cleaned_df['Area'].apply(lambda x: x if is_country(x) else None)
coahd_cleaned_df_new['Category'] = coahd_cleaned_df['Area'].apply(lambda x: x if is_category(x) else None)
coahd_cleaned_df_new
```
